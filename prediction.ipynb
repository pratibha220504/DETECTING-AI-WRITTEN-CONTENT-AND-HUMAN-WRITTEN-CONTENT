```python
# =========================================
# 1. Import Libraries
# =========================================

import joblib
import re


# =========================================
# 2. Load Saved Model and Vectorizer
# =========================================

model = joblib.load("logistic_model.sav")
vectorizer = joblib.load("tfidf_vectorizer.pkl")

print("Model Loaded Successfully")


# =========================================
# 3. Text Cleaning Function
# =========================================

def clean_text(text):

    text = str(text).lower()

    text = re.sub(r'http\S+', '', text)      # remove URLs
    text = re.sub(r'[^a-zA-Z ]', '', text)   # remove special characters
    text = re.sub(r'\s+', ' ', text)         # remove extra spaces

    return text.strip()


# =========================================
# 4. Prediction Function
# =========================================

def predict_text(input_text):

    cleaned = clean_text(input_text)

    vector = vectorizer.transform([cleaned])

    prediction = model.predict(vector)[0]

    return prediction


```

```python
# =========================================
# 5. Test with New Record
# =========================================

new_text = "We investigate the hydrological response of watersheds to wildfire, analyzing the changes in peak flow and flooding potential."
result = predict_text(new_text)

print("\nInput Text:")
print(new_text)

print("\nPrediction:")
print(result)
```
